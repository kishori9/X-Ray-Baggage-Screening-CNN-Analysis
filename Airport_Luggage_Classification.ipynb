## Converting RGB images from Google to greyscale

* Reference: https://stackoverflow.com/questions/47087528/converting-images-in-a-folder-to-grayscale-using-python-and-opencv-and-writing-i

```
import cv2 #OpenCV library
import os,glob #retrieve files/pathnames matching a specified pattern
from os import listdir,makedirs #make and get the list of all files and directories in a specified directory
from os.path import isfile,join #check whether the specified path is an existing regular file or not
path = '/content/drive/Shareddrives/OPIM 5509 Group 5/Project 2/Data/RGB non-restricted' # Source Folder
dstpath = '/content/drive/Shareddrives/OPIM 5509 Group 5/Project 2/Data/Non-restricted' # Destination Folder

#throw exception if the destination folder already exists 
#if the directory alredy exists, images will be written to the same folder
try:
    makedirs(dstpath)
except:
    print ("Directory already exist, images will be written in same folder")

#get a list of all the files in the directory 
files = list(filter(lambda f: isfile(join(path,f)), listdir(path)))

#converting all the images to greyscale on by one
for image in files:
    try:
        img = cv2.imread(os.path.join(path,image)) #read an image from source folder
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #convert it to greyscale
        dstPath = join(dstpath,image) #store greyscaled image to destination folder
        cv2.imwrite(dstPath,gray)
    except:
        print ("{} is not converted".format(image)) #throw exception if the image could not be converted
for fil in glob.glob("*.jpg"): #to covert jpeg files to greyscale
    try:
        image = cv2.imread(fil) 
        gray_image = cv2.cvtColor(os.path.join(path,image), cv2.COLOR_BGR2GRAY) # convert to greyscale
        cv2.imwrite(os.path.join(dstpath,fil),gray_image)
    except:
        print('{} is not converted')
```

# 3. Data Preprocessing

* Currently our images are in jpeg format
* We use the following steps to get the images into the network 
   * Read the image files
   * Decode the jpeg content to RGB(Red,Green,Blue) grids of pixels
   * Convert these into floating point tensors
   * Rescale the pixel values between 0 and 255 to (0,1) interval since neural networks prefer to deal with small input values.
* Keras utilities take care of these steps automatically. Keras has a module with image processing helper tools, located at keras.preprocessing.image . It contains the class ImageDataGenerator which allows to quickly set up Python generators that can automatically turn image files on disk into batches of pre-processed tensors.
* We rescale images in train_dir and test_dir and store them into train_generator and validation_generator

```
from keras.preprocessing.image import ImageDataGenerator

# All images will be rescaled by 1./255
train_datagen = ImageDataGenerator(rescale=1./255)
validation_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        # This is the target directory
        train_dir,
        # All images will be resized to 150x150
        target_size=(150, 150),
        color_mode='grayscale', # define the color mode parameter in BW as flow_from_directory will convert them back to RGB as default
        batch_size=20,
        # Since we use binary_crossentropy loss, we need binary labels
        class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(150, 150),
        color_mode='grayscale', # define the color mode parameter in BW as flow_from_directory will convert them back to RGB as default
        batch_size=20,
        class_mode='binary')
```

# 4. Model 

## Model 1: Convolution Neural Network
* We will be using Sequential API to build the model
  * Hidden units in Convolution Layer 1: 32
  * Hidden units in Convolution Layer 2: 64
  * Hidden units in Convolution Layer 3: 128
  * Hidden units in Convolution Layer 4: 128
  * Dense Layers: 512

* Compile the Model using
  * Optimizer: Adam
  * Loss function: binary_crossentropy
  * Additional Metrics: Accuracy

* We will make a new variable called 'history' to evaluate the learning curves. 
  * epochs of 6
  * steps per epoch: 20

* Activation functions used:
  * relu
  * sigmoid (since it is a binary classification)




We have four convolution and max pooling layers with relu activation function. In addition, it has Dense Layer with 512 units and one output layer with sigmoid function.


**Input Layer :** The input layer reads the x-ray scanned baggage images. Our input has 150 * 150 pixels of black and white images. This means that the image has 150 rows and 150 columns. Since we have a black and white image, the channel list is 1. Tha batch is always None in all of the layers.
* Trainable Parameters: There are no trainable in input layers
* Output Size: (None, 150, 150, 1)

**Convolutional Layer- Conv2d :**
  * Filter size (M x N) = 3 x 3
  * Channels (L) : 1
  * Bias (B) : 1 ( Bias is alywas 1)
  * Filter = 32 feature maps.
  * Kernel Size : (3, 3)

The trainable parameters are calculated by using the above formula, (3 x 3 x 1 + 1) x 32 = **320**. From the model, we get that there are 32 channels, each training image generates a 32 dimensional feature map. We now calculate the output size (So - (M - 1)) = (150 - (3- 1)) = 148. The input image 150 x 150 is slimmed down to 148 x 148.

The first Convolutional Layer Conv2d gives 896 trainable parameters and an output size of (None, 148, 148, 32).

**Maxpooling Layer- max_pooling2d :**
There are no trainable parameters in a pooling layer. Here in the pooling layer, we are aggressively down sampling with 2 x 2 window that will never overlap. Also, to get the new dimensions, we divide the axis by filter size, i.e; 2 in this case. 
New Dimension = 148/2 = 74
Coming to the filter, we are inheriting it from the previous convolutional layer (Conv2d) i.e; 16. The single output from that layer is just squished from 148 x 148 to 74 x 74. 
The output size of max_pooling2d is (None, 74, 74, 32)






**Convolutional Layer- Conv2d_1 :**
  * Filter size (M x N) = 3 x 3
  * Channels (L) : 32 (from previous step)
  * Bias (B) : 1 ( Bias is alywas 1)
  * Filter = 64 feature maps.
  * Kernel Size : (3,3)

The trainable parameters are calculated by using the above formula, (3 x 3 x 32 + 1) x 64 = **18,496**. We are pealing one of these feature maps off and then doing convolution on it with 64 feature maps. From the model, we get that there are 64 channels. We now calculate the output size (So - (M - 1)) = (74 - (3- 1)) = 72. The input image 74 x 74 is slimmed down to 72 x 72.

The second Convolutional Layer Conv2d_1 gives 18,496 trainable parameters and an output size of (None, 72, 72, 64).

**Maxpooling Layer- max_pooling2d_1 :**
There are no trainable parameters in a pooling layer. Here in the pooling layer, Again, we are aggressively down sampling with 2 x 2 window that will never overlap. Also, to get the new dimensions, we divide the axis by filter size, i.e; 2 in this case. 
New Dimension = 72/2 = 36
We are inheriting the filter from the previous convolutional layer (Conv2d_1) i.e; 64 and we do pooling with a 3 x 3 window. The single output from that layer is just squished from 72 x 72 to 36 x 36. 
The output size of max_pooling2d is (None, 36, 36, 64)

**Convolutional Layer- Conv2d_2 :**
  * Filter size (M x N) = 3 x 3
  * Channels (L) : 64 (from previous step)
  * Bias (B) : 1 ( Bias is always 1)
  * Filter = 128 feature maps.
  * Kernel Size : (3,3)

The trainable parameters are calculated by using the above formula, (3 x 3 x 64 + 1) x 128 = **73,856**. We are pealing one of these feature maps off and then doing convolution on it with 42 feature maps. From the model, we get that there are 128 channels. We now calculate the output size (So - (M - 1)) = (36 - (3- 1)) = 34. The input image 36 x 36 is slimmed down to 34 x 34.

The thrid Convolutional Layer Conv2d_2 gives 73,856 trainable parameters and an output size of (None, 34, 34, 128).

**Maxpooling Layer- max_pooling2d_2 :**
There are no trainable parameters in a pooling layer. Here in the pooling layer, we are aggressively down sampling with 3 x 3 window that will never overlap. Also, to get the new dimensions, we divide the axis by filter size, i.e; 2 in this case. 
New Dimension = 34/2 = 17
We are inheriting the filter from the previous convolutional layer (Conv2d_2) i.e; 128 and we do pooling with a 2 x 2 window. The single output from that layer is just squished from 34 x 34 to 17 x 17. 
The output size of max_pooling2d is (None, 17, 17, 128)

**Convolutional Layer- Conv2d_3 :**
  * Filter size (M x N) = 3 x 3
  * Channels (L) : 128 (from previous step)
  * Bias (B) : 1 ( Bias is alywas 1)
  * Filter = 128 feature maps.
  * Kernel Size : (3,3)

The trainable parameters are calculated by using the above formula, (3 x 3 x 128 + 1) x 128 = **147,584**. We are pealing one of these feature maps off and then doing convolution on it with 128 feature maps. From the model, we get that there are 128 channels. We now calculate the output size (So - (M - 1)) = (17 - (3- 1)) = 15. The input image 17 x 17 is slimmed down to 15 x 15.

The thrid Convolutional Layer Conv2d_2 gives 147,584 trainable parameters and an output size of (None, 15, 15, 128).

**Maxpooling Layer- max_pooling2d_3 :**
There are no trainable parameters in a pooling layer. Here in the pooling layer, we are aggressively down sampling with 3 x 3 window that will never overlap. Also, to get the new dimensions, we divide the axis by filter size, i.e; 2 in this case. 
New Dimension = 15/2 = 7 
We are inheriting the filter from the previous convolutional layer (Conv2d_2) i.e; 128 and we do pooling with a 2 x 2 window. The single output from that layer is just squished from 15 x 15 to 7 x 7. 
The output size of max_pooling2d is (None, 7, 7, 128)

**Flatten Layer (flatten) :**
There are no trainable parameters in a flatten layer as it is going from a matrix to a vector. The pooled feature maps are flattened as an input to a dense nueral network. The output size is calculated from the output in the previous step. Rows x Columns x Channels from max_pooling2d_3 layer : **7 x 7 x 128 = 6272**.
The output size of Flatten Layer is (None, 6272)

**Dense Layer (dense (512)):**
This is a fully connected layer. We get the number of hidden layers 512 from the code.
The total number of trainable parameters for this layer is calculated using the formula: (Input Size + Bias) x Number of Hidden Layers of Current Layer =  **(6272 + 1) x 512 = 3,211,776**. There are 3,211,776 trainable parameters and the output size is just the size of the layer = (None, 512)

**Output Layer (dense_1):**
We get the number of hidden layers = 1 from the code.
The total number of trainable parameters for this layer is calculated using the formula : (Input Size + Bias) x Number of Hidden Layers of Current Layer =  **(512 + 1) x 1 = 513**. There are 513 trainable parameters and the output size is just the size of the layer = (None, 1).
A sigmoid acivation function is used here.

**Total Number of Parameters:**
Finally, the total number of trainable parameters are calculated by adding up the trainable parameters from all the layers.

## Model 2: Modeling with Data Augmentation

In Model 2, we are using Data Augmentation approach to generate more training data from existing training samples to avoid overfitting. We are using following techniques of Data Augmentation to increase the size of our dataset 
  * `rotation_range` is a value in degrees (0-180), a range within which to randomly rotate pictures.
  * `width_shift` and `height_shift` are ranges (as a fraction of total width or height) within which to randomly translate pictures 
vertically or horizontally.
  * `shear_range` is for randomly applying shearing transformations.
  * `zoom_range` is for randomly zooming inside pictures.
  * `horizontal_flip` is for randomly flipping half of the images horizontally
  * `fill_mode` is the strategy used for filling in newly created pixels, which can appear after a rotation or a width/height shift.

The goal is that at training time, our model does not see the exact same picture twice. This will help the model get exposed to more aspects of the data and generalize better.
We use the same model architecture as model 1 but with data augmentation using ImageGenerator



```
# Data Augmentation to generate more images for training data from existing training sample by rotating, flipping, zooming the images.
datagen = ImageDataGenerator(
      rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.4,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')
```

## Model 3 & 4: Pre-trained Network - Feature Extraction

Since our restricted and non-restricted images dataset is small, we are going to use VGG16 to leverage a pre-trained network. A pre-trained network is a saved network that was previously trained on a large dataset. Feature extraction is one way to use a pretrained network. It consists of using the representations learned by a previous network to extract interesting features from new samples. These features are run through a new classifier which is trained from scratch. Feature extraction consists of taking the convolutional base of a previously trained network, running the new data through it, and training a new classifier on top of the output. It will get to an accuracy of 90% to 96%.

```
from keras.applications import VGG16

conv_base = VGG16(weights='imagenet', # weights from imagenet that is pre-trained on ImageNet
                  include_top=False, # our dense layer will be trained specifically to our small dataset problem - restricted vs. non-restricted
                  input_shape=(150, 150, 3)) # the shape of our data from our data generator
```

## Model 3: Fast Feature Extraction Without Data Augmentation

We simply run `ImageDataGenerator` to extract images as Numpy arrays and their labels. We will extract features from these images simply by calling the `predict` method of the `conv_base` model. Using this data as input to a standalone densely connected classifier. This solution is very fast and cheap to run as it only requires running the convolutional base once for every input image, and the convolutional base is by far the most expensive part of the pipeline. For the same reason, this technique will not allow to use data augmentation.

```
# add our data from the path
path = '/content/drive/Shareddrives/OPIM 5509 Group 5/Project 2/Data/'
train_dir = path + "train"
test_dir = path + "test"
```

```
from datetime import datetime
startTime = datetime.now()

# initialize our generator
datagen = ImageDataGenerator(rescale=1./255)
batch_size = 20

def extract_features(directory, sample_count):
    features = np.zeros(shape=(sample_count, 4, 4, 512)) # this is the shape of final feature map from our VGG16
    labels = np.zeros(shape=(sample_count))
    generator = datagen.flow_from_directory(
        directory,
        target_size=(150, 150),
        batch_size=batch_size,
        class_mode='binary')
    i = 1
    for inputs_batch, labels_batch in generator:
        features_batch = conv_base.predict(inputs_batch)
        features[i * batch_size : (i + 1) * batch_size] = features_batch
        labels[i * batch_size : (i + 1) * batch_size] = labels_batch
        i += 1
        if i * batch_size >= sample_count:
            # the generators yield data indefinitely in a loop
            # we must `break` after every image has been seen once
            break
    return features, labels

train_features, train_labels = extract_features(train_dir, 420)
print(datetime.now() - startTime)
test_features, test_labels = extract_features(test_dir, 280)
print(datetime.now() - startTime)
```

```
# X features for modeling
train_features = np.reshape(train_features, (420, 4 * 4 * 512))
test_features = np.reshape(test_features, (280, 4 * 4 * 512))
```

```
# define our densely-connected classifier
# the use of dropout for regularization
# train it on the data and labels

model = models.Sequential()
model.add(layers.Dense(256, activation='relu', input_dim=4 * 4 * 512))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer=optimizers.RMSprop(lr=2e-5),
              loss='binary_crossentropy',
              metrics=['acc'])

# create history variable to evaluate learning curve
history = model.fit(train_features, train_labels,
                    epochs=20, # number of epochs
                    batch_size=20, # number of batch size
                    validation_data=(test_features, test_labels))
```

Training is very fast in this method as we only have to deal with two `Dense` layers.

```
import matplotlib.pyplot as plt

# take a look at the loss and accuracy curves

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'bo', label='Training acc') # show the training accuracy in blue dots
plt.plot(epochs, val_acc, 'orange', label='Validation acc') # show the validation accuracy in orange line
plt.title('Training and validation accuracy') # name the title of the plot
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss') # show the training loss in blue dots
plt.plot(epochs, val_loss, 'orange', label='Validation loss') # show the validation loss in orange line
plt.title('Training and validation loss') # name the title of the plot
plt.legend()

plt.show() # show the plot
```

## Model 5: Pre-trained Network - Fine-Tuning

Fine-tuning consists of unfreezing a few of the top layers of a frozen model base used for feature extraction, and jointly training both the newly added part of the model and these top layers. It slightly adjusts the more abstract representations of the model being reused to make them more relevant for the problem. This will get a final accuracy of 97%.

It is necessary to freeze the convolution base of VGG16 to be able to train a randomly initialized classifier on top. It is only possible to fine-tune the top layers of the convolutional base once the classifier on top has already been trained. If the classifier is not already trained, the error signal propagating through the network during training will be too large and the representations previously learned by the layers being fine-tuned will be destroyed.

```
# the first 3 steps is completed in feature extraction
# proceed with the 4th step to unfreeze the conv_base
# then freeze individual layers inside of it
# check the convolutional base again
conv_base.summary()
# we will fine-tune the last 3 convolutional layers
# all layers up until block4_pool should be frozen
# the layers block5_conv1, block5_conv2 and block5_conv3 should be trainable
```

We will fine-tune the last three convolutional layers, including `block5_conv1`, `block5_conv2`, and `block5_conv3` should be trainable. Earlier layers in the convolutional base encode more-generic, reusable featuers, whereas layers higher up encode more-specialized featuers. It is more useful to fine-tune the more specialized features as these are the ones that need to be repurposed on the new problem. There would be fast-decreasing returns in fine-tuning lower layers.

Also, the more parameters we training increases the risks of overfitting. Since convolutional base has 15 million parameters, it would be risky to attempt to train it on our small dataset.

```
# all layers up until blcok4_pool will be frozen
# only the layers related to block5_conv1, block5_conv2, block5_conv3 will be trainable
conv_base.trainable = True

set_trainable = False
for layer in conv_base.layers:
    if layer.name == 'block5_conv1':
        set_trainable = True
    if set_trainable:
        layer.trainable = True
    else:
        layer.trainable = False
```

We are using a low learning rate to limit the magnitude of the modifications we make to the representations of the three layers we are fine-tuning.

```
# proceed with fine-tuning
model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-5),
              metrics=['acc'])

# create history variable to evaluate learning curve
history = model.fit_generator(
      train_generator,
      steps_per_epoch=10, # number of steps per epoch
      epochs=20, # number of epochs
      validation_data=validation_generator,
      validation_steps=10)
```

In order to increase the readability, we smooth the plot by replacing every loss and accuracy with exponential moving averages of these quantities.

```
# since our original plot is noisy, we smooth the plot to increase the readability
def smooth_curve(points, factor=0.8):
  smoothed_points = []
  for point in points:
    if smoothed_points:
      previous = smoothed_points[-1]
      smoothed_points.append(previous * factor + point * (1 - factor))
    else:
      smoothed_points.append(point)
  return smoothed_points

plt.plot(epochs,
         smooth_curve(acc), 'bo', label='Smoothed training acc') # show the smoothed training accuracy in blue dots
plt.plot(epochs,
         smooth_curve(val_acc), 'orange', label='Smoothed validation acc') # show the smoothed validation accuracy in orange line
plt.title('Training and validation accuracy') # name the title
plt.legend()

plt.figure()

plt.plot(epochs,
         smooth_curve(loss), 'bo', label='Smoothed training loss') # show the smoothed training loss in blue dots
plt.plot(epochs,
         smooth_curve(val_loss), 'orange', label='Smoothed validation loss') # show the smoothed validation loss in orange line
plt.title('Training and validation loss') # name the title
plt.legend()

plt.show() # show the plot
```
